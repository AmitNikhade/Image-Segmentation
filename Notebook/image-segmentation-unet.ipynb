Preprocesing Data

```python
X_train = np.zeros((len(images), 256, 256, 3), dtype=np.float32)
y_train = np.zeros((len(masks), 256, 256, 1), dtype=np.bool)

for n, img in tqdm(enumerate(images)):
    image = tf.keras.preprocessing.image.load_img(f'../input/segmentation-full-body-mads-dataset/segmentation_full_body_mads_dataset_1192_img/images/'+img, target_size=(IMG_HEIGHT, IMG_WIDTH))
    input_arr = tf.keras.preprocessing.image.img_to_array(image)/ 255 
    image = tf.keras.preprocessing.image.array_to_img(input_arr, ).resize((256, 256))
    image = np.array(image)
    print(image.shape)
    X_train[n] = image
    
# for n, img in tqdm(enumerate(masks)):
#     image = tf.keras.preprocessing.image.load_img(f'../input/segmentation-full-body-mads-dataset/segmentation_full_body_mads_dataset_1192_img/masks/'+img, target_size=(IMG_HEIGHT, IMG_WIDTH), color_mode="grayscale")
#     input_arr = tf.keras.preprocessing.image.img_to_array(image)/ 255 
#     image = tf.keras.preprocessing.image.array_to_img(input_arr).resize((256, 256))
#     image = np.array(image)
#     print(image.shape)
#     y_train[n] = image[:, :, np.newaxis]
for n, img in tqdm(enumerate(masks)):
    im = cv2.imread("../input/segmentation-full-body-mads-dataset/segmentation_full_body_mads_dataset_1192_img/masks/"+img)
    im = np.array(cv2.resize(im,(256,256)))[:,:,:1]
    y_train[n] = im
#     im.shape
```

```python
X_train = np.append(X_train, [np.fliplr(x) for x in X_train], axis=0)
y_train = np.append(y_train, [np.fliplr(x) for x in y_train], axis=0)
```

Model

```python
# UNET MODEL

# Convolutional Block

def Conv_block(n, inp):
    x = tf.keras.layers.Conv2D(n, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(inp)
    x = tf.keras.layers.Dropout(0.1)(x)
    xc = tf.keras.layers.Conv2D(n, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(x)
    x = tf.keras.layers.MaxPooling2D((2, 2))(xc)
    return x, xc

# Upsampling
def upsample(n ,inp, c):
    
    x = tf.keras.layers.Conv2DTranspose(n, (2, 2), strides=(2, 2), padding='same')(inp)
    x = tf.keras.layers.concatenate([x, c])
    x = tf.keras.layers.Conv2D(n, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(x)
    x = tf.keras.layers.Dropout(0.2)(x)
    x = tf.keras.layers.Conv2D(n, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(x)
    return x
    
# fully connected layers

def fcn(inp):
    x1,xc1 = Conv_block(16, inp)
    x2, xc2 = Conv_block(16*2,x1)
    x3, xc3 = Conv_block(16*4,x2)
    x4, xc4 = Conv_block(16*8,x3)

    x = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(x4)
    x = tf.keras.layers.Dropout(0.3)(x)
    x = tf.keras.layers.Conv2D(256, (3, 3), activation='relu', kernel_initializer='he_normal', padding='same')(x)

    us1 = upsample(16*8,x,xc4)
    us2 = upsample(16*4,us1,xc3)
    us3 = upsample(16*2,us2,xc2)
    us4 = upsample(16,us3,xc1)
    return us4


inputs = tf.keras.layers.Input((256, 256, 3))
inp = tf.keras.layers.Lambda(lambda x: x / 255)(inputs)
net = fcn(inp)
outputs = tf.keras.layers.Conv2D(1, (1, 1), activation='sigmoid')(net)
model = tf.keras.Model(inputs=[inputs], outputs=[outputs])

# Compile Model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

Training

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

# defining callbacks
early_stopping = EarlyStopping(patience=10, verbose=1)
model_checkpoint = ModelCheckpoint("./keras.model", save_best_only=True, verbose=1)
reduce_lr = ReduceLROnPlateau(factor=0.1, patience=4, min_lr=0.00001, verbose=1)

epochs = 20
batch_size = 64

# Start training
history = model.fit(X_train, y_train,
                    validation_split = 0.1,
                    epochs=epochs,
                    batch_size=batch_size,
                    callbacks=[early_stopping, model_checkpoint, reduce_lr],shuffle=True)
```

Performance

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'r', label='Training accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
plt.title('Training and validation accuracy')
plt.legend(loc=0)
plt.figure()


plt.show()
```
